```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import glob
import zipfile
```

```python
def unzip(source_file, dest_path):
    with zipfile.ZipFile(source_file, 'r') as zf:
        zipInfo = zf.infolist()
        for member in zipInfo:
            try:
                print(member.filename.encode('cp437').decode('euc-kr', 'ignore'))
                member.filename = member.filename.encode('cp437').decode('euc-kr', 'ignore')
                zf.extract(member, dest_path)
            except:
                print(source_file)
                raise Exception('what?!')
```

```python
#G:\tmp_project\img_train\image\갈비탕\Q1\Q1.zip

base_dir = "/content/drive/MyDrive/YANG/unzip"
img_folders = glob.glob(base_dir+"/img*")

for train_val in img_folders:
    food_folders = glob.glob(train_val +"/image/*")
    for food in tqdm(food_folders):
        quantity_folders = glob.glob(food +"/*")
        quantity_zips = glob.glob(food +"/**/*.zip")
        for i in range(len(quantity_folders)):
            print(f"unzip: {quantity_zips[i]}")
            unzip(quantity_zips[i], quantity_folders[i])
```
