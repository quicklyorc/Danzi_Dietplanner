```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import glob
import xml.etree.ElementTree as ET
```

```python
#G:\tmp_project\img_train\image\갈비탕\Q1\.jpg

base_dir = "/content/drive/MyDrive/YANG/unzip"
img_folders = glob.glob(base_dir+"/img*")
label_folders = glob.glob(base_dir+"/label*")
```

```python
for train_val in label_folders:
    food_folders = glob.glob(train_val +"/xml/*")
    for food in tqdm(food_folders):
        foodname = food.split("/")[-1]
        quantity_xmls = glob.glob(food +"/**/*.xml")
        food_label = pd.DataFrame(columns = ["filename","filepath","quantity","size_w","size_h","c_xmin","c_ymin","c_xmax","c_ymax","d_xmin","d_ymin","d_xmax","d_ymax","f_xmin","f_ymin","f_xmax","f_ymax"])

        for xml in quantity_xmls:

            #xml = xml.replace("\\","/")

            # XML 파일 읽기
            tree = ET.parse(xml)
            root = tree.getroot()

            # 자식 엘리먼트 순회
            filename = root.find("filename").text
            wsize = root.find('size/width').text
            hsize = root.find('size/height').text

#             print(filename)
            torv = train_val.split("_")[-1]
            path = xml.split("/")
            path[6] = f"img_{torv}"
            path[7] = "image"
            quantity = path[-2]
            path[-1] = filename
            filepath = "/".join(path)
            filelabel = [filename,filepath,quantity,wsize,hsize]

            for obj in root.findall('object'):
                if obj.find("name").text != "ref_spoon" and obj.find("name").text != "ref_fork":
                    bndbox = obj.find('bndbox')
                    xmin = bndbox.find('xmin').text
                    ymin = bndbox.find('ymin').text
                    xmax = bndbox.find('xmax').text
                    ymax = bndbox.find('ymax').text

                    filelabel += [xmin, ymin, xmax, ymax]

            food_label.loc[len(food_label)] = filelabel
        food_label.to_csv(food+f"/{foodname}.csv", index=False)
```
