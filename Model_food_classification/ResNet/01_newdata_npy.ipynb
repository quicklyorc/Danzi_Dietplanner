```python
# new100 npy 변경
import numpy as np
import pandas as pd
import csv
import cv2
from tqdm import tqdm
import os
import gc

base_path = '/content/drive/MyDrive/지윤/food100'
data_dirs = [os.path.join(base_path, sub_dir) for sub_dir in ['1','2','3','4']]

# 클래스 레이블을 저장할 딕셔너리
class_indices = {}
current_index = 0

img_size = (224, 224)

data = []
target = []
for data_dir in data_dirs:
    for class_name in os.listdir(data_dir):
        class_path = os.path.join(data_dir, class_name)
        if os.path.isdir(class_path): #True
            if class_name not in class_indices:
                class_indices[class_name] = current_index
                current_index += 1
            class_index = class_indices[class_name]
            image_files = sorted(os.listdir(class_path))
            for img_name in tqdm(image_files[:200]):
                img_path = os.path.join(class_path, img_name)
                img = cv2.imread(img_path)
                if img is not None:
                    img = cv2.resize(img, img_size)
                    data.append(img)
                    target.append(class_index)

# numpy 배열로 변환
data = np.array(data) / 255.0
data16 = data.astype(np.float16)
target = np.array(target)

# 데이터와 타겟을 파일로 저장

np.save('/content/drive/MyDrive/지윤/new100/npy/data_001.npy', data16)
np.save('/content/drive/MyDrive/지윤/new100/npy/target_001.npy', target)

# 메모리 해제
del data
del data16
del target
gc.collect()

df = pd.DataFrame(list(class_indices.items()), columns=['Class Name', 'Class Index'])
csv_path = '/content/drive/MyDrive/지윤/new100/npy/class_001.csv'
df.to_csv(csv_path, index=False)

print("개별 파일이 성공적으로 저장되었습니다!")
```
