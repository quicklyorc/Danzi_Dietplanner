## json과 이미지 갯수 다른 음식명 차이 확인 및 삭제

```python
import glob
import os

# 압축 파일 경로와 해제할 경로 설정
unzip_file_path = '/content/drive/MyDrive/food_data/Training/unzip'

# ZIP 파일들의 경로 리스트를 얻기
unzip_files = glob.glob(os.path.join(unzip_file_path, '*'))
num_files = sorted(unzip_files)

# 차이가 있는 요소들을 저장할 리스트
differences = []

for file_path in num_files[35:36]:
    print("\n확인할 폴더: " + file_path)
    img_path = os.path.join(file_path, 'image')
    json_path = os.path.join(file_path, 'json')

    img_file_path = glob.glob(os.path.join(img_path, '*'))
    img_sort_files = sorted(img_file_path)

    print("음식 폴더의 수:", len(img_file_path))

    for file in img_sort_files:
        imgs = glob.glob(os.path.join(file, '*'))
        food_name = file.split('/')[-1]
        food_json = food_name + " json"
        print(f"\t{food_name}의 이미지 수:", len(imgs))

        json_sort_files = os.path.join(json_path, food_json)
        jsons = glob.glob(os.path.join(json_sort_files, '*'))
        print(f"\t{food_name}의 json 수:", len(jsons))

        if len(imgs) != len(jsons):
            print(f"\tWarning: {food_name}의 이미지와 json 파일의 수가 다릅니다.")
            img_names = {os.path.splitext(os.path.basename(img))[0] for img in imgs}
            json_names = {os.path.splitext(os.path.basename(json))[0] for json in jsons}

            # 이미지와 JSON 파일 이름 차이를 계산
            img_not_in_json = img_names - json_names
            json_not_in_img = json_names - img_names

            # 차이 목록에 추가
            differences.append({
                'food_name': food_name,
                'img_not_in_json': list(img_not_in_json),
                'json_not_in_img': list(json_not_in_img)
            })

# 차이가 있는 요소들을 출력 및 파일 제거
for diff in differences:
    print(f"\n{diff['food_name']}의 이미지와 JSON 파일의 차이:")
    print(f"\t이미지에 있지만 JSON에 없는 파일: {diff['img_not_in_json']}")
    print(f"\tJSON에 있지만 이미지에 없는 파일: {diff['json_not_in_img']}")
    print(f"(총 이미지에 있지만 JSON에 없는 파일: {len(diff['img_not_in_json'])}개)")
    print(f"(총 JSON에 있지만 이미지에 없는 파일: {len(diff['json_not_in_img'])}개)")

    # 이미지 파일 제거
    for img_name in diff['img_not_in_json']:
        img_file = os.path.join(img_path, diff['food_name'], f"{img_name}.jpg")
        if os.path.exists(img_file):
            os.remove(img_file)
            print(f"삭제된 이미지 파일: {img_file}")

    # JSON 파일 제거
    for json_name in diff['json_not_in_img']:
        json_file = os.path.join(json_path, diff['food_name'] + " json", f"{json_name}.json")
        if os.path.exists(json_file):
            os.remove(json_file)
            print(f"삭제된 JSON 파일: {json_file}")
```

## json zip 파일에 들어가있는 파일 갯수 확인

```python
import pandas as pd

def searchzip(source_file):
    folderlist = []
    filelist = []
    aa = False
    with zipfile.ZipFile(source_file, 'r') as zf:
        zipInfo = zf.infolist()
        for member in zipInfo:
            try:
                filename = member.filename.encode('cp437').decode('euc-kr', 'ignore')
#                 print(filename[-1]=="/")
                if filename[-1] == "/" and aa == False:
                    if len(filelist) == 0:
                        folderlist.append(filename[:-1])
                        aa = True

                    else:
                        folderlist.append(len(filelist))
                        folderlist.append(filename[:-1])
                        filelist=[]
                        aa = True

                elif filename[-1] == "/" and aa == True:
                    folderlist.append(len(filelist))
                    folderlist.append(filename[:-1])
                else:
                    filelist.append(filename)
                    aa = False
            except:
                print(source_file)
                raise Exception('what?!')
    folderlist.append(len(filelist))

    print(folderlist)

    f_dict = {folderlist[i]: folderlist[i + 1] for i in range(0, len(folderlist), 2)}

    return f_dict


# 압축 파일 경로와 해제할 경로 설정
zip_files = ['/content/drive/MyDrive/food_data/Training/json/[라벨]음식104_Tra_json.zip',]


files_dict = {}
for zip_file in zip_files: ### 여기 수정 하시면 됩니당!
    print('start: ' + zip_file)
    tmp_dict = searchzip(zip_file)
    files_dict.update(tmp_dict)

files_df = pd.DataFrame([files_dict])
```
